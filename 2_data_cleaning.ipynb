```python
# Reagenge columns
df_data = df_data[['crypto_id', 'date', 'open_time', 'close_time', 'open', 'close', 'high', 'low', 'volume', 'number_of_trades']]
df_data
```

```python
# Make a copy of the data
df = df_data.copy()
df
```

```python
df.info()
```

```python
# Change column date to datetime
df['date'] = pd.to_datetime(df['date'])
```

```python
df = df.groupby(['date', 'crypto_id']).agg({
    'open': 'first',
    'close': 'last',
    'high':'max',
    'low': 'min',
    'volume': 'sum',
    'number_of_trades': 'sum'}).reset_index()
df
```

```python
# Sort by date and crypto_id to ensure correct calculations
df = df.sort_values(by=['crypto_id', 'date'])
```

```python
# Calculate diference betewen open and close price and to a column diff_oc
df['diff_oc'] = df['close'] - df['open']
```

```python
# Calculate daily return (% change in close price)
df['return'] = ((df['close'] - df['open']) / df['open']) * 100
```

```python
# Calculate volatility (% price fluctuation)
df['volatility'] = ((df['high'] - df['low']) / df['open']) * 100
```

```python
# Calculate volume change percentage from the previous day
df['vol_change'] = df.groupby('crypto_id')['volume'].pct_change() * 100
```

```python
# Calculate moving averages (5-day and 10-day) for closing price
df['ma_5'] = df.groupby('crypto_id')['close'].transform(lambda x: x.rolling(window=5).mean())
df['ma_10'] = df.groupby('crypto_id')['close'].transform(lambda x: x.rolling(window=10).mean())
```

```python
# Create classification target labels
df['price_movement'] = (df['close'] > df['open']).astype(int)  # 1 if price went up, 0 otherwise
df.head()
```

```python
df
```

```python
# check for nan values
df.isna().sum()
```

```python
def fill_missing_values(df):
    for column in df.columns:
        if df[column].dtype == 'float64' or df[column].dtype == 'int64':
            df[column].fillna(0, inplace=True)
        elif df[column].dtype == 'object':
            df[column].fillna('unknown', inplace=True)
        elif df[column].dtype == 'datetime64[ns]':
            df[column].fillna(pd.Timestamp('1900-01-01'), inplace=True)

# Call the function
fill_missing_values(df)
```

```python
# check for nan values
df.isna().sum()
```

```python
df.head()
```

```python
# Select relevant features for correlation analysis
features = ['open', 'close', 'high', 'low', 'volume', 'number_of_trades',
            'diff_oc', 'return', 'volatility', 'vol_change', 'ma_5', 'ma_10', 'price_movement']

# Create a new dataframe with selected features
df_features = df[features].copy()

# Convert 'price_movement' to categorical
df_features['price_movement'] = df_features['price_movement'].astype('category')

# Calculate correlation matrix
corr_matrix = df_features.corr()

# Display the correlation matrix
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm', cbar=True)
plt.title('Correlation Matrix of Crypto Features')
plt.show()



```

```python
# Check correlaion between 'volatility' and other features
corr_volatility = df_features.corr()['volatility'].sort_values(ascending=False)
print("Correlation with Volatility:")
print(corr_volatility)
```
