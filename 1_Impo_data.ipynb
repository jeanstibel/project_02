```python
# Define the folder path and the list of cryptocurrencies
folder_path = '/Users/jeanstibel/Documents/AI Classes/Projects/project_02/Data'
cryptos = ['AAVEUSDT', 'ADAUSDT', 'ALGOUSDT', 'ATOMUSDT', 'AVAXUSDT', 'BATUSDT', 'BNBUSDT', 'BTCUSDT', 'BCHUSDT', 
           'CAKEUSDT', 'CHZUSDT', 'CRVUSDT', 'DASHUSDT', 'DEXEUSDT', 'DOGEUSDT', 'DOTUSDT', 'ENAUSDT', 'ENJUSDT', 'EOSUSDT', 'ETCUSDT', 
           'ETHUSDT', 'FILUSDT', 'GRTUSDT', 'HBARUSDT', 'ICPUSDT', 'IOSTUSDT', 'IOTAUSDT', 'LINKUSDT', 'LTCUSDT', 
           'MANAUSDT', 'NEARUSDT', 'NEOUSDT', 'QTUMUSDT', 'RVNUSDT', 'SANDUSDT', 'SHIBUSDT', 'SOLUSDT', 
           'STMXUSDT', 'SUSHIUSDT', 'TFUELUSDT', 'THETAUSDT', 'TIAUSDT', 'TRXUSDT', 'UNIUSDT', 'VETUSDT',
           'XLMUSDT', 'XRPUSDT', 'XTZUSDT', 'ZECUSDT', 'ZILUSDT']

# Load the data and add the crypto_id column
data_frames = {}
for crypto in cryptos:
    file_path = os.path.join(folder_path, crypto, f'{crypto}.csv')
    data_frames[crypto] = pd.read_csv(file_path)
    data_frames[crypto]['crypto_id'] = crypto
```

```python
# Concatenate all dataframes into one
all_data_df = pd.concat(data_frames.values(), ignore_index=True)
all_data_df
```

```python
# Rearanging the columns
all_data_df = all_data_df[['crypto_id', 'timestamp',  'close_time', 'open',  'close', 'high', 'low', 'volume', 'quote_asset_volume',
                       'number_of_trades', 'taker_buy_base_asset_volume', 'taker_buy_quote_asset_volume',
                       'ignore']]
```

```python
# drop columns 'ignore', 'quote_asset_volume', 'taker_buy_quote_asset_volume', 'taker_buy_base_asset_volume'
all_data_df  = all_data_df .drop(columns=['ignore', 'quote_asset_volume', 'taker_buy_quote_asset_volume', 'taker_buy_base_asset_volume'])
```

```python
# From column 'crypto_id' remove -USD
all_data_df['crypto_id'] = all_data_df['crypto_id'].str.replace('USDT', '')
all_data_df
```
